# üå¶ Weather Dataset Analysis

We will explore:
- Dataset structure
- Unique weather conditions
- Data cleaning and grouping
- Filtered insights (e.g., fog, snow)
- Visualizations using Seaborn & Matplotlib


**üì¶ Import Required Libraries**

We begin by importing the required libraries:
- `pandas` for data handling  
- `matplotlib.pyplot` and `seaborn` for visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("‚úÖ Libraries imported successfully: pandas, matplotlib, seaborn")
```

**üìÇ Load the Weather Dataset**

We will now load the dataset using `pd.read_csv()`.

```python
data = pd.read_csv(r"C:\Users\CON11052\Downloads\Project+1+-+Weather+Dataset.csv")
print("‚úÖ Dataset loaded successfully!")
```

**üîç Initial Exploration**

To get a quick overview, we display the first and last 5 rows of the dataset.

```python
print("üîπ First 5 rows:")
print(data.head())

print("\nüîπ Last 5 rows:")
print(data.tail())
```

**üìè Dataset Structure**

We examine the dataset's shape, index, column names, and data types.

```python
print("üîπ Shape:", data.shape)
print("\nüîπ Index:", data.index)
print("\nüîπ Columns:", data.columns)
print("\nüîπ Data Types:\n", data.dtypes)
```

**üî¢ Unique Values and Non-Null Counts**

We check for unique values and counts of non-null data in each column.

```python
print("üîπ Unique Weather Conditions:\n", data['Weather'].unique())
print("üîπ Unique per column:\n", data.nunique())
print("üîπ Non-null counts:\n", data.count())
```

**üìä Weather Condition Frequencies**

We count how often each unique weather condition appears.

```python
print(data['Weather'].value_counts())
```

**üß† Dataset Info and Missing Values**

We use `.info()` and null-checking functions to understand data quality.

```python
print(data.info())
print("\nMissing values per column:\n", data.isnull().sum())
print("\nNot null values:\n", data.notnull().sum())
```

**‚úèÔ∏è Rename Columns**

To improve clarity, we rename the 'Weather' column to 'Weather Condition'.

```python
data.rename(columns={'Weather': 'Weather Condition'}, inplace=True)
print("Column 'Weather' renamed to 'Weather Condition'")
print("Updated columns:", data.columns.tolist())
```

**üìà Descriptive Statistics**

We compute mean, standard deviation, and variance for key columns.

```python
print("Mean Visibility:", data['Visibility_km'].mean())
print("Std Dev of Pressure:", data['Press_kPa'].std())
print("Variance of Humidity:", data['Rel Hum_%'].var())
```

**‚ùÑÔ∏è Filter - Snow**

We filter all rows that include "Snow" in the weather condition.

```python
data[data['Weather Condition'].str.contains('Snow', case=False)]
```

**üí® Wind Speed > 24 and Visibility = 25**

We filter the data based on two conditions using logical AND.

```python
data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)]
```

**üë• Grouping by Weather Condition**

We group data by weather condition to compute mean, maximum and minimum values.

```python
print(data.groupby('Weather Condition').mean(numeric_only=True))
print(data.groupby('Weather Condition').max(numeric_only=True))
print(data.groupby('Weather Condition').min(numeric_only=True))
```

**üåÅ Filter - Fog**

We filter all rows where weather condition is exactly "Fog".

```python
data[data['Weather Condition'] == 'Fog']
```

**üå§ Clear or Visibility > 40**

We filter rows where weather is Clear or visibility exceeds 40.

```python
data[(data['Weather Condition'] == 'Clear') | (data['Visibility_km'] > 40)]
```

**üå§ Clear + Humidity > 50 OR Visibility > 40**

We use combined conditions with proper operator precedence.

```python
data[((data['Weather Condition'] == 'Clear') & (data['Rel Hum_%'] > 50)) | (data['Visibility_km'] > 40)]
```

**üìÖ Add Time Features**

We extract month and hour from the Date/Time column.

```python
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
data['Month'] = data['Date/Time'].dt.month
data['Hour'] = data['Date/Time'].dt.hour

print("'Date/Time' column converted to datetime format.")
print("'Month' column created:", data['Month'].unique())
print("'Hour' column created:", data['Hour'].unique())
```

**üìä Top 15 Weather Conditions (Bar Plot)**

We plot a horizontal bar chart showing the most frequent weather types.

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Weather Condition', data=data, order=data['Weather Condition'].value_counts().index[:15])
plt.title("Top 15 Weather Conditions")
plt.tight_layout()
plt.show()
```

**üå° Temperature Over Time (Line Plot)**

We visualize temperature changes over time.

```python
plt.figure(figsize=(14, 5))
sns.lineplot(x='Date/Time', y='Temp_C', data=data)
plt.title("Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (¬∞C)")
plt.tight_layout()
plt.show()
```

**üìÜ Monthly Average Temperature**

We compute and plot average monthly temperatures.

```python
monthly_avg = data.groupby('Month')['Temp_C'].mean()
plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_avg.index, y=monthly_avg.values, hue=monthly_avg.index, palette='coolwarm', legend=False)
plt.title("Average Temperature by Month")
plt.xlabel("Month")
plt.ylabel("Temperature (¬∞C)")
plt.tight_layout()
plt.show()
```

**üí® Wind Speed Distribution (Histogram)**

We create a histogram with a KDE curve for wind speed distribution.

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['Wind Speed_km/h'], kde=True, bins=30, color='skyblue')
plt.title("Wind Speed Distribution")
plt.xlabel("Wind Speed (km/h)")
plt.tight_layout()
plt.show()
```

**üëÅ Visibility vs Wind Speed (Scatter Plot)**

We analyze correlation visually between visibility and wind speed.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Visibility_km', y='Wind Speed_km/h', hue='Weather Condition', data=data, legend=False)
plt.title("Visibility vs Wind Speed")
plt.xlabel("Visibility (km)")
plt.ylabel("Wind Speed (km/h)")
plt.tight_layout()
plt.show()
```

**üß™ Correlation Heatmap**

We create a heatmap showing the correlation matrix for numeric columns.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

**üïí Temperature by Time Period**

We categorize each row into morning, afternoon, evening, or night based on the hour, and display temperature stats per period.

```python
def get_time_period(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

data['Time Period'] = data['Hour'].apply(get_time_period)
print(data.groupby('Time Period')['Temp_C'].describe())
```

**üëÅ Average Visibility for Top 10 Weather Conditions**

We calculate and visualize the average visibility for the 10 most frequent weather conditions.

```python
top_conditions = data['Weather Condition'].value_counts().head(10).index
avg_visibility = data[data['Weather Condition'].isin(top_conditions)].groupby('Weather Condition')['Visibility_km'].mean().sort_values()

plt.figure(figsize=(10,5))
sns.barplot(x=avg_visibility.values, y=avg_visibility.index, hue=avg_visibility.index, palette='viridis', legend=False)
plt.title("Average Visibility for Top 10 Weather Conditions")
plt.xlabel("Visibility (km)")
plt.ylabel("Weather Condition")
plt.tight_layout()
plt.show()
```

**üíß Distribution of Relative Humidity**

We explore the distribution and spread of relative humidity using descriptive stats and a boxplot.

```python
print("Summary statistics for Relative Humidity:")
print(data['Rel Hum_%'].describe())

plt.figure(figsize=(10,5))
sns.boxplot(x=data['Rel Hum_%'], color='skyblue')
plt.title("Boxplot of Relative Humidity")
plt.xlabel("Relative Humidity (%)")
plt.tight_layout()
plt.show()
```

**üåû Day vs Night Temperature Comparison**

We split the data into day and night based on hour, and compare temperature statistics.

```python
data['Day/Night'] = data['Hour'].apply(lambda x: 'Day' if 6 <= x < 18 else 'Night')
print(data.groupby('Day/Night')['Temp_C'].agg(['mean', 'min', 'max']))

sns.violinplot(x='Day/Night', y='Temp_C', data=data, hue='Day/Night', palette='pastel', legend=False)
plt.title("Temperature Distribution: Day vs Night")
plt.tight_layout()
plt.show()
```

**üìà Pressure Trends Across Months**

We analyze the average atmospheric pressure over each month.

```python
monthly_pressure = data.groupby('Month')['Press_kPa'].mean()

plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_pressure.index, y=monthly_pressure.values, marker='o', color='crimson')
plt.title("Average Atmospheric Pressure by Month")
plt.xlabel("Month")
plt.ylabel("Pressure (kPa)")
plt.tight_layout()
plt.show()
```

**üìä Correlation Between Temperature, Pressure & Humidity**

We calculate and visualize the correlation among core weather attributes.

```python
subset = data[['Temp_C', 'Press_kPa', 'Rel Hum_%']]
correlation = subset.corr()
print("Correlation matrix:\n", correlation)

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation: Temperature, Pressure, Humidity")
plt.tight_layout()
plt.show()
```

**üìÜ Most Common Weather Condition per Month**

We identify the most frequently observed weather condition in each month.

```python
common_weather_month = data.groupby('Month')['Weather Condition'].agg(lambda x: x.value_counts().idxmax())
print("Most common weather by month:\n", common_weather_month)
```

**üå°Ô∏è Daily Temperature Range**

We calculate the daily temperature range (max - min) and identify the days with the highest variation.

```python
data['Date'] = data['Date/Time'].dt.date
daily_range = data.groupby('Date')['Temp_C'].agg(['min', 'max'])
daily_range['Range'] = daily_range['max'] - daily_range['min']

print("Top 5 days with highest temperature fluctuation:")
print(daily_range.sort_values(by='Range', ascending=False).head())
```

**üå§Ô∏è Hourly Average Temperature by Weather Condition**

We calculate average temperature per hour for each of the top 5 weather conditions.

```python
top5_weather = data['Weather Condition'].value_counts().head(5).index
hourly_temp = data[data['Weather Condition'].isin(top5_weather)].groupby(['Hour', 'Weather Condition'])['Temp_C'].mean().unstack()

hourly_temp.plot(figsize=(12, 6), cmap='coolwarm')
plt.title("Hourly Average Temperature by Weather Condition")
plt.xlabel("Hour of Day")
plt.ylabel("Average Temperature (¬∞C)")
plt.legend(title='Weather Condition')
plt.tight_layout()
plt.show()
```

**üí® Average Wind Speed by Time of Day**

We calculate how average wind speed varies across different times (morning, afternoon, evening, night).

```python
wind_by_period = data.groupby('Time Period')['Wind Speed_km/h'].mean().reindex(['Morning', 'Afternoon', 'Evening', 'Night'])

plt.figure(figsize=(8,5))
sns.barplot(
    x=wind_by_period.index,
    y=wind_by_period.values,
    hue=wind_by_period.index,
    palette='Blues',
    legend=False
)
plt.title("Average Wind Speed by Time Period")
plt.ylabel("Wind Speed (km/h)")
plt.tight_layout()
plt.show()
```

**üåßÔ∏è Rain/Snow Conditions and Humidity Levels**

We analyze humidity during rain or snow to identify environmental conditions.

```python
rain_snow = data[data['Weather Condition'].str.contains('Rain|Snow', case=False)]
print("Average humidity during Rain/Snow:", rain_snow['Rel Hum_%'].mean())
```

**üå´Ô∏è Frequency of Low Visibility Events**

We count how often visibility was very low (‚â§5 km).

```python
low_visibility = data[data['Visibility_km'] <= 5]
print(f"Number of low visibility events (‚â§5 km): {low_visibility.shape[0]}")
```

**üìâ Coldest and Hottest Days of the Year**

We find the dates with the minimum and maximum recorded temperature.

```python
min_temp_day = data.loc[data['Temp_C'].idxmin()]
max_temp_day = data.loc[data['Temp_C'].idxmax()]

print("Coldest day:", min_temp_day['Date/Time'], "Temperature:", min_temp_day['Temp_C'], "¬∞C")
print("Hottest day:", max_temp_day['Date/Time'], "Temperature:", max_temp_day['Temp_C'], "¬∞C")
```

**üî• Extreme Weather Frequency**

We calculate how often extreme weather conditions like "Thunderstorm", "Snow", or "Freezing" occurred.

```python
extreme_keywords = ['thunderstorm', 'snow', 'freezing', 'blizzard', 'hail', 'ice pellets']
extreme_weather = data[data['Weather Condition'].str.lower().str.contains('|'.join(extreme_keywords))]

print("Total extreme weather records:", extreme_weather.shape[0])
print("Extreme weather breakdown:")
print(extreme_weather['Weather Condition'].value_counts())
```

**üìå Temperature vs Humidity Correlation by Time Period**

We explore how the relationship between temperature and humidity changes by part of the day (Morning, Afternoon, Evening, Night).

```python
correlations = data.groupby('Time Period')[['Temp_C', 'Rel Hum_%']].corr().iloc[0::2, -1]
print("Temperature vs Humidity correlation by time period:\n")
print(correlations)
```

**üßä Sub-Zero Temperature Frequency by Month**

We count how many hours had temperatures below 0¬∞C, grouped by month.

```python
sub_zero = data[data['Temp_C'] < 0]
sub_zero_by_month = sub_zero['Month'].value_counts().sort_index()

plt.figure(figsize=(8,4))
sns.barplot(
    x=sub_zero_by_month.index, 
    y=sub_zero_by_month.values, 
    hue=sub_zero_by_month.index, 
    palette='cool', 
    legend=False
)
plt.title("Sub-Zero Temperature Events by Month")
plt.xlabel("Month")
plt.ylabel("Number of Hours")
plt.tight_layout()
plt.show()
```

**üîÅ Repeating Weather Patterns**

We identify the most frequent back-to-back weather condition transitions (e.g., from Clear to Cloudy).

```python
weather_shift = data['Weather Condition'].shift()
transitions = weather_shift + " ‚ûù " + data['Weather Condition']
most_common_transitions = transitions.value_counts().head(10)

print("Most common weather transitions:")
print(most_common_transitions)
```

**üåÄ Rapid Wind Speed Changes**

We detect hours where wind speed changed drastically compared to the previous hour.

```python
data['Wind_Change'] = data['Wind Speed_km/h'].diff().abs()
rapid_wind = data[data['Wind_Change'] > 20]
print(f"Hours with wind speed change > 20 km/h: {rapid_wind.shape[0]}")
print(rapid_wind[['Date/Time', 'Wind Speed_km/h', 'Wind_Change']].head())
```

**üåô Nighttime Visibility Distribution**

We plot the visibility distribution specifically during nighttime hours.

```python
night_data = data[data['Time Period'] == 'Night']

plt.figure(figsize=(10,5))
sns.histplot(night_data['Visibility_km'], kde=True, bins=25, color='purple')
plt.title("Visibility Distribution at Night")
plt.xlabel("Visibility (km)")
plt.tight_layout()
plt.show()
```

**üîç Weather Description Length Analysis**

We analyze the number of words used in different weather condition descriptions to check detail richness.

```python
data['Weather_Desc_Length'] = data['Weather Condition'].apply(lambda x: len(x.split()))
desc_stats = data.groupby('Weather Condition')['Weather_Desc_Length'].mean().sort_values(ascending=False)

print("Top 10 most descriptive weather condition labels (by word count):")
print(desc_stats.head(10))
```
